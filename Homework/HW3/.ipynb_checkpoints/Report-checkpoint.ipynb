```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
masked = [78, 50, 40, 33, 30, 27, 24, 23.5, 23, 22.8]
non_masked = [88, 55, 40, 37, 32, 27, 24, 23.5, 23, 20.8]

fig, ax = plt.subplots(1, figsize=(10, 8))
ax.plot(masked, label='Masked Attention', color='blue')
ax.plot(non_masked, label='Vanilla Attention', color='red')
ax.set_xticks(range(10))
ax.set_xlabel('Epochs')
ax.set_ylabel('Perplexity')
ax.legend()
ax.set_title('Comparison of Attention models')
```
